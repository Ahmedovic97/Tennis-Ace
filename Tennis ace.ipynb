```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('tennis_stats.csv')
df.head()
```

## Exploratory analysis 


```python
plt.scatter(df[['FirstServe']], df[['Winnings']], alpha=0.4)
plt.show()
```

```python
plt.scatter(df[['FirstServePointsWon']], df[['Winnings']], alpha=0.4)
plt.show()
```

## single feature linear regressions

```python
feature1 = df[['BreakPointsConverted']]

winnings = df[['Winnings']]

feature1_train, feature1_test, winnings_train, winnings_test = train_test_split(feature1, winnings, train_size=0.8)

model = LinearRegression()
model.fit(feature1_train, winnings_train)
print('Predicting winnings with BreakPoints converted:', model.score(feature1_test, winnings_test))

prediction1 = model.predict(feature1_test)
plt.scatter(winnings_test, prediction1, alpha=0.4)
plt.xlabel('Actual Winnings', fontsize=12)
plt.ylabel('Predicted Winnings', fontsize=12)
plt.title('Actual Winnings vs Predicted Winnings using Breakpoint Converted Feature')
plt.show()
```

## multiple feature linear regressions

```python
features1and2 = df[['BreakPointsConverted', 'FirstServePointsWon']]
winnings = df[['Winnings']]

features1and2_train, features1and2_test, winnings_train, winnings_test = train_test_split(features1and2, winnings, train_size= 0.8)

model = LinearRegression()
model.fit(features1and2, winnings)
print('Predicting winnings with FirstServe and FirstServePointsWon:', model.score(features1and2_test, winnings_test))

two_features_prediction = model.predict(features1and2_test)

plt.scatter(winnings_test, two_features_prediction, alpha=0.4)
plt.xlabel('Actual Winnings', fontsize=12)
plt.ylabel('Predicted Winnings', fontsize=12)
plt.title('Actual Winnings vs Predicted Winnings using BreakPointsConverted and FirstServePointsWon Feature')
plt.show()
```
